```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Carregar el dataset
df = pd.read_csv("airquality.csv")

# Mostrar les 10 primeres files
df.head(10)

```

```python
# Longitud del dataset
print(f"La longitud del dataset és: {len(df)}")
```

```python
# Información sobre el dataset
df.info()
```

```python
# Obtener la temperatura máxima
max_temp = df["Temp"].max()
print(f"La temperatura máxima registrada es: {max_temp}")
```

```python
# Filtrar la fila donde está la temperatura máxima
max_temp_row = df[df["Temp"] == max_temp]

# Obtener el mes y el día de la temperatura máxima
mes_max = max_temp_row["Month"].values[0]
dia_max = max_temp_row["Day"].values[0]

print(f"La temperatura máxima se registró el día {dia_max} del mes {mes_max}.")
```

```python
# Crear un gráfico scatter entre Wind y Solar.R
plt.figure(figsize=(8,6))
sns.scatterplot(x=df["Wind"], y=df["Solar.R"])
plt.xlabel("Velocidad del Viento")
plt.ylabel("Radiación Solar")
plt.title("Relación entre Velocidad del Viento y Radiación Solar")
plt.show()
```

```python
# Calcular la temperatura media por mes
temp_mensual = df.groupby("Month")["Temp"].mean()

# Crear el gráfico de barras
plt.figure(figsize=(8,6))
temp_mensual.plot(kind="bar", color="skyblue")
plt.xlabel("Mes")
plt.ylabel("Temperatura Media")
plt.title("Temperatura Media por Mes")
plt.show()
```

```python
# Crear un gráfico scatter entre Ozone y Temp
plt.figure(figsize=(8,6))
sns.scatterplot(x=df["Ozone"], y=df["Temp"])
plt.xlabel("Nivel de Ozono")
plt.ylabel("Temperatura")
plt.title("Relación entre Ozono y Temperatura")
plt.show()
```
