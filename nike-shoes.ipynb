```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Cargar el dataset
df = pd.read_csv("nike_shoes_sales.csv")
```

```python
# Mostrar informaci√≥n general
df.info()
```

```python
df.head()
```

```python
num_products = len(df)
print("N√∫mero total de productos:", num_products)
```

```python
# Precio m√°ximo, m√≠nimo y promedio
max_price = df["sale_price"].max()
min_price = df["sale_price"].min()
avg_price = df["sale_price"].mean()
print(f"Precio m√°ximo: {max_price}, Precio m√≠nimo: {min_price}, Precio promedio: {avg_price:.2f}")
```

```python
# Relaci√≥n entre descuento y precio de venta
sns.scatterplot(data=df, x="discount", y="sale_price")
plt.title("Relaci√≥n entre descuento y precio de venta")
plt.show()
```

```python
# Relaci√≥n entre rating y n√∫mero de rese√±as
sns.scatterplot(data=df, x="rating", y="reviews")
plt.title("Relaci√≥n entre calificaci√≥n y n√∫mero de rese√±as")
plt.show()
```
