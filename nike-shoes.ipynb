```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Cargar el dataset
df = pd.read_csv("nike_shoes_sales.csv")
```

```python
# Mostrar información general
df.info()
```

```python
df.head()
```

```python
num_products = len(df)
print("Número total de productos:", num_products)
```

```python
# Precio máximo, mínimo y promedio
max_price = df["sale_price"].max()
min_price = df["sale_price"].min()
avg_price = df["sale_price"].mean()
print(f"Precio máximo: {max_price}, Precio mínimo: {min_price}, Precio promedio: {avg_price:.2f}")
```

```python
# Relación entre descuento y precio de venta
sns.scatterplot(data=df, x="discount", y="sale_price")
plt.title("Relación entre descuento y precio de venta")
plt.show()
```

```python
# Relación entre rating y número de reseñas
sns.scatterplot(data=df, x="rating", y="reviews")
plt.title("Relación entre calificación y número de reseñas")
plt.show()
```
